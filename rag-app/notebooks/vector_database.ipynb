```python
import chromadb
```

```python
client = chromadb.Client()
```

```python
collection = client.create_collection(name="News", get_or_create=True)
```

```python
collection.add(
      documents=[
            "Apple repprted its quarterly earnings today",
            "Apple has a lot of Vitamin A"
      ],
      ids = ["id1", "id2"]
)
```

```python
objects = collection.peek()
objects
```

```python
len(objects['embeddings'][0])
```

```python
results = collection.query(
      query_texts=["The flagship phones launched by Samsung are too good. They had recorded all time sales last year."],
      n_results = 2
)
results
```

```python
results = collection.query(
      query_texts=["I had a fruit custard yesterday."],
      n_results = 2
)
results
```

```python
from datetime import datetime
```

```python
collection = client.create_collection(name="reviews",
      metadata={
            "description":"Product Reviews",
            "created": str(datetime.now())
      }
)
```

```python
collection = client.get_collection("reviews")
```

```python
collection.add(
    documents=[
        "The delivery product delivery was fast and the product quality is excellent",
        "I was not able to increase TV's brightness So I returned it back",
        "The Shoes I ordered were too small. Sizing is inaccurate",
        "Great customer support. Resolved my issue in minutes."
    ],
    ids=["r1","r2","r3","r4"],
    metadatas=[
        {"product_category":"electronics", "rating":5},
        {"product_category":"electronics", "rating":2},
        {"product_category":"apparel", "rating":3},
        {"product_category":"services", "rating":4}
    ]
)
```

```python
collection.count()
```

```python
collection.peek()
```

```python
collection.query(
      query_texts=["fast shipping"],
      n_results=2
)
```

# UPDATE

```python
collection.update(
      documents=["The shoes I ordered were replaced with a larger size. Fits perfectly now!"],
      ids=['r3'],
      metadatas=[
            {"product_category":"apparel","rating":5}
      ]
)
```

```python
query_results = collection.query(
      query_texts=["shoe size"],
      n_results=2,
      where={"product_category":"apparel"}
)
```

# DELETE

```python
collection.delete(ids=["r1", "r2"])
```

```python
client.delete_collection(name='reviews')
```
